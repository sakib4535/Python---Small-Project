```python
import matplotlib.pyplot as plt
import numpy as np
lx, ly = 0.020, 0.06
x = np.arange(-4.0, 4.0, lx)
y = np.arange(-4.0, 4.0, ly)
X, Y = np.meshgrid(x,y)
extend = np.min(x), np.max(x), np.min(y), np.max(y)

ps = np.add.outer(range(8), range(8)) % 2
plt.imshow(ps, cmap="binary_r", interpolation="nearest", extent=extend, alpha=1)

def chess(x,y):
    return(1 -x / 2 + x ** 5 + y ** 6) * np.exp(-(x ** 2 + y ** 2))
c1 = chess(X,Y)
plt.imshow(c1, alpha=0, interpolation="bilinear", extent=extend)
plt.title("Chess Board For You")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return np.sin(np.sqrt(x **2 + y ** 2))
x = np.linspace(-8, 8, 80)
y = np.linspace(-8, 8, 80)
x, y =np.meshgrid(x,y)
z = f(x,y)
figure = plt.figure()
ax =plt.axes(projection='3d')
ax.contour3D(x,y,z,50, cmap='binary')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
```
